```python
# https://arxiv.org/pdf/1707.02574.pdf

import numpy as np
import random

class CategoricalVariable:
    def __init__(self,probs):
        self.probs=probs/sum(probs)
        self.num_choices=len(probs)
    def draw_element(self):
        value=random.random()
        for i in range(self.num_choices):
            if (value<=self.probs[i]):
                return i
            else:
                value = value-self.probs[i]
        return (self.num_choices-1)
    def draw_and_update(self,delta):
        drawn_element=self.draw_element()
        for i in range(self.num_choices):
            self.probs[i]*=(1-delta)
        self.probs[drawn_element]+=delta
        return drawn_element
    
cv1=CategoricalVariable(np.array([.5,.5]))
# independent trials
for i in range(20):
    print(cv1.draw_element())
# the probabilities change based on 
for i in range(20):
    print(cv1.draw_and_update(.2))
    print(cv1.probs)
```
